#### RecursionError: maximum recursion depth exceeded while calling a Python object
#### 재귀의 최대 깊이를 초과하면 발생하는 오류 코드

* 해당 오류 때문에 재귀함수를 허용할 경우 종료 조건을 꼭 명시해야한다.
* 명시하지 않으면 함수가 무한 호출될 수 있기 때문이다.
* 해결 방법 : import sys 
* sys.setrecursionlimit(상수)

```python
def recursive_function(i):
    #100번째 출력했을 때 종료되도록 종료 조건을 적기
    if i == 100:
        return 
    print(i, '번째 재귀함수에서', i+1, '번째 재귀함수를 호출합니다')
    recursive_function(i+1)
    print(i, '번째 재귀함수를 종료합니다')
recursive_function(1)
```

```python
#재귀 함수의 대표적인 예시 : 팩토리얼 문제(Factorial)

#반복문으로 구현한 n!
def factorial_iterative(n):
    result = 1
    #1부터 n까지의 수를 차례대로 곱하기
    for i in range(1, n+1):
        result *=i
    return result

def factorial_resursive(n):
    if n<=1: #n이 1이하인 경우 1을 반환
        return 1
    #n! = n * (n-1)!
    return n * factorial_resursive(n-1)

print(factorial_iterative(5))
print(factorial_resursive(5))
    
```

코드가 상대적으로 간결해진 이유 : 수확의 재귀식(점화식)을 그대로 소스코드로 옮겼기 때문임(?)  
점화식은 특정한 함수를 자신보다 더 작은 변수에 대한 함수와의 관계로 표현한 것을 의미함  
이러한 점화식은 이후 다이나믹 프로그래밍으로 연계됨
