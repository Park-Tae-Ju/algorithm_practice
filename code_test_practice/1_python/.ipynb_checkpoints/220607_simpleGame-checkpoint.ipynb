## 오늘 만들 것 : 기억력 게임
- 입력된 숫자를 순서대로 기억해서 입력하는 게임


### 시나리오
1. 먼저 랜덤하게 숫자를 일정량 유저에게 보여준다
2. 보여준 후 해당 숫자를 가린다
3. 가린 숫자를 사용자에게 맞추라고 유도한다
4. 잘 맞췄으면 난이도를 올린다

### 완성된 시나리오
1. 먼저 랜덤하게 숫자를 일정량 유저에게 보여준다
2. 숫자 하나당 1초 보여준 후 해당 숫자를 가린다
3. 가린 숫자를 사용자에게 맞추라고 지시한다. 단, 한번에 정답을 제출할 것
4. 잘 맞췄으면 난이도를 올린다(+2)

### 평가

```python
#1. 랜덤하게 숫자를 일정량 유저에게 보여준다
#2. 보여준 후 해당 숫자를 가린다

#랜덤한 숫자를 일정량 가지고 있는 리스트를 만든다.
import random
import time
level = 5
quit = 'n'
count = 0

while quit != 'y':
    question = [random.randrange(1,10) for _ in range(level)]

    print('5초 뒤 게임이 시작됩니다.')
    for i in range(5,0,-1):
        time.sleep(1)
        print(i, end = '\r')
    
    print('시작합니다')
    
    for i in question:
        print(i, end = '\r') # 이전 출력문 삭제 후 출력
        time.sleep(1) #1초간 딜레이
    print(' ', end = '\r')

    #3.사용자에게 맞추라고 유도한다
    #각 숫자를 띄어쓰기로 구분해서 입력하라고 유도
    print('보여진 숫자를 순서대로 입력하세요')
    answer = list(map(int,input('각 숫자는 띄어쓰기로 구분해주세요 : ').split()))

    #현재 상황 : 문제에 대한 답안을 한꺼번에 입력받아서 확인하고자 하는 상황
    #3-1. 맞으면 맞았다고 문제 전체에 대한 각각의 답안을 돌려줄 것인지
    checker = []
    for i in range(len(question)):
        if answer[i] != question[i]: #문제가 틀렸을 경우
            checker = checker + [False]
        else:
            checker = checker + [True]
    print('문제 중 맞춘 숫자입니다')
    print(question)
    print(checker)        

    #3-2. 어디가 틀렸는지는 가르쳐주지 않고, 그냥 틀렸다라고만 진행할 것인지 : 함수화
    # for i in range(len(question)):
    #     if answer[i] != question[i]: #문제가 틀렸을 경우
    #         print('틀렸습니다!')
    #         break


    #종료 만들기
    while True:
        quit = input('종료하시겠습니까?(y/n) : ')
        if quit not in 'yn':
            print('다시 입력해주세요')
        else:
            break
            
    #4.잘 맞췄으면 난이도를 올린다
    #4-1. 잘 맞췄으면 = checker안에 있는게 전부 True = all 함수 사용하거나 in 함수 사용   
    if all(checker):
        level += 2
        count += 1
else:
    print('종료되었습니다. 맞춘 문제는 {}개, 최종 난이도는 {}입니다.'.format(count,level))  
```

```python
import random
import time

def make_question(level):
    return [random.randrange(1,10) for _ in range(level)]

def startAndShow(showing, question):
    print('{}초 뒤 게임이 시작됩니다.'.format(showing))
    for i in range(5,0,-1):
        print(i, end = '\r')
        time.sleep(1)
    print('시작합니다')
    for i in question:
        print(i, end = '\r') 
        time.sleep(1)
    print(' ', end = '\r')

def submit_answer():
    print('보여진 숫자를 순서대로 입력하세요')
    answer =  list(map(int,input('각 숫자는 띄어쓰기로 구분해주세요 : ').split()))
    return answer
    
def checking_answer(question):
    checker = []
    for i in range(len(question)):
        if answer[i] != question[i]:
            checker = checker + [False]
        else:
            checker = checker + [True]
            
    #3-2. 어디가 틀렸는지는 가르쳐주지 않고, 그냥 틀렸다라고만 진행할 것인지 : 함수화
    # for i in range(len(question)):
    #     if answer[i] != question[i]: #문제가 틀렸을 경우
    #         print('틀렸습니다!')
    #         break         
    return checker

def upgrade_level(checker, level, upper = 2):
    if all(checker): #전부 맞췄으면 난이도를 올리고, 아니면 내린다
        level += upper
        check = True
    else:
        if level >= 3:
            level -= upper
        else:
            level = 1
        check = False
    return level, check
    

level = 5
quit = 'n'
count = 0
showing = 5

while quit != 'y':   
    question = make_question(level)
    startAndShow(showing, question)
    answer = submit_answer()
    checker = checking_answer(question)
    
    print('문제 중 맞춘 문제입니다')
    print(question)
    print(checker)        

    while True:
        quit = input('종료하시겠습니까?(y/n) : ')
        if quit not in 'yn':
            print('다시 입력해주세요')
        else:
            level, check = upgrade_level(checker, level, upper = 2)
            if check == True:
                count += 1
            break

else:
    print('종료되었습니다. 맞춘 문제는 {}개, 최종 난이도는 {}입니다.'.format(count,len(question)))  
```
