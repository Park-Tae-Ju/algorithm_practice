```python
#웹과 통신을 하기 위한 라이브러리
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen

#데이터를 저장하기 위한 파일 관련 라이브러리
import os
```

```python
#웹툰 1장을 가져오기
url = 'https://comic.naver.com/webtoon/detail?titleId=687915&no=281&weekday=mon'
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.88 Safari/537.36'
         }

#통신 요청(request)하기
webToon = requests.get(url, headers = header)
soup = BeautifulSoup(webToon.text, 'html.parser') #웹페이지의 태그들을 txt 형태로 가져오기
wtViewer = soup.find('div', {'class' : 'wt_viewer'})
imgUrl = wtViewer.find_all('img')[3]['src']
imgRequest = requests.get(imgUrl, headers = header)

#이미지 파일을 저장하기 위한 경로 만들기
savePath = 'C:/Users/Administrator/parktaeju/평일_11(A,1)/머신러닝/wtData/testWt/'
try:
    if not os.path.exists(savePath):
        os.makedirs(savePath)
except OSError:
    print('error : already created directory')
    
# 파일 경로에 이미지 파일을 저장하기
imgNum = 0
with open(savePath + str(imgNum) + '.jpg', 'wb') as f:
    f.write(imgRequest.content)
f.close()
```
