# 데이터 생성 및 편집하기

```python
data = np.random.randint(10, size = (5,8))
df = pd.DataFrame(data)

#원본 데이터를 복제해서 수정용 데이터를 작성
df1 = df.copy()
```

```python
df1['rowSum'] = df1.sum(axis=1)
df1
```

```python
df1.loc['colSum', :] = df1.sum(axis=0)
df1
```

```python
#평균으로 새로운 열,행 추가
df1['rowMean'] = df.mean(axis = 1)
df1.loc['colMean', :] = df.mean(axis = 0)
df1
```

```python
#데이터 갯수가 같도록 구간을 나누는 방법(qcut)
import pandas as pd
data = [x for x in range(1000)]
label = [x for x in range(15)]
dataQcut = pd.qcut(data, 15, labels = label)
dataQcut.value_counts()
```

```python
#데이터 합치기(병합)
#merge : 2개 이상의 데이터프레임의 key를 기준으로 합침 / key : 열, 인덱스

df1 = pd.DataFrame({
    '학급 번호' : [1,2,3,4,5,6],
    '이름' : ['철수', '영희', '진구',' 짱구', '퉁퉁이', '비실이']})

df2 = pd.DataFrame({
    '학급 번호' : [1,2,3,4,5,6],
    '상점' : [3,6,9,1,5,7]
})

# pd.concat([df1,df2], axis = 1 )
df = pd.merge(df1,df2)
```

```python
#키 값이 전부 있지 않아도 merge가 가능하다
#outer join

df3 = pd.DataFrame({
    '학급 번호' : [1,4,5,6],
    '시험 성적' : [64,87,93,56]
})
pd.merge(df, df3, how = 'outer')
```

```python
pd.merge(df, df3, how = 'right')
```

```python
pd.merge(df, df3, how = 'left')
```

```python
pd.merge(df, df3)
```

```python
pd.merge(df3, df)
```

```python
#공통된 것이 여러개 있을 경우 원하는 칼럼을 key로 해서 합치기
pd.merge(df,df3, on = '학급 번호')
```

```python
#pivot table : sub dataset
train.pivot_table('Pclass', 'Age')
```

```python
#groupby : 먼저 나온 데이터를 기준으로 뒤에 나온 데이터를 집계
#기술 통계와 연관해서 씀 : sum, mean, median, count, var, min, max 등등

#성별 기준 나이의 중위값
trainGroup = train.groupby(train.Sex)['Age'].median()

#성별 기준 나이의 평균
trainGroup = train.groupby(train.Sex)['Age'].mean()

#성별 데이터를 등급 및 성별 기준으로 집계(갯수를 센 것)
trainGroup = train.groupby([train.Pclass, train.Sex,train.Survived])['Survived'].count()
trainGroup
```
