# 데이터 분석 프로세스
- 요건 정의 - 모델링 - 검증 및 테스트 - 적용

### 1. 요건 정의

#### 구체적인 내용을 실무 담당자와 협의하는 업무

분석방안도출
- 비즈니스 이슈로부터 분석 요건을 도출한다
- 분석 요건의 조건은 문제를 해결했을 때, 이것이 수익으로 증명이 되는지 여부
- 현황 파악 및 간단한 기초분석, 기존 조사 자료 참고 등을 수행해서 방향성을 제시한다
- 요구사항 추출, 분석, 명세, 검증, 유지보수 등을 수행한다.(IEEE의 요구 공학 프로세스)
- 필요한 능력 : 커뮤니케이션, 단어에 대한 사전 정의 및 협의, 개인정보 확인 등

수행방안 설계
- 구체적인 수행방안을 설계
- 탐색적 분석을 실행 : 가설을 수립하고, 어떠한 분석을 이후에 수행할지 기초를 세우기 위한 선행작업
- 기존에 분석된 자료들을 확인할 것
- 분석 계획서 작성 및 deadline 확인 등

요건 확정
- 분석 관련 항목을 기준으로 이 작업이 진행되어야하는 의미가 있는지를 확정 짓는 것
- 해당 사항을 문서화, 분석 요건과 관련된 적합성, 일정 등의 제약을 종합해서 업무 범위를 조정한다.
- 데이터 분석 요건을 명시화하여 문서화하는 것이 중요하며, 명시화된 요건을 변경 시, 반드시 공식 절차에 따라서 진행한다.


### 2. 모델링

#### 유의미한 변수를 추출하여 개발된 모델들의 성능을 평가하고, 최적의 모델을 선정하는 업무

모델링 마트 설계와 구축
- 모델링을 위한 데이터를 미리 준비하는 단계
- DB에서 값을 가져와서 모델링을 할 수도 있음(SPSS 등)
- 클라이언트와의 커뮤니케이션을 통해 원천 데이터의 명확한 이해가 필요
- 데이터의 획득, 수정, 확정이 지연되는 경우가 있으므로 해당 과정에 대한 기간 확정을 통한 작업 수행에 대한 명시
- 데이터 정제 과정의 하나인 데이터 요약, 파생변수 도출, 변수 확대의 접근 방법을 고려

탐색적 분석과 유의변수 도출
- EDA(exploratory data analysis): 데이터 마이닝에 해당하는 부분이며, 이를 통해 데이터 내부에 있는 구체적인 사실 관계를 발견하기 위해 시행하는 사항
- 시각화 정보를 제시할 때, 사실 및 통찰과 관련된 정보를 제공할 수 있도록 그래프를 디자인
- 시각화를 통해 높은 유의성을 보유한 변수들을 식별한다
- 만약, 이 과정에서 받은 데이터의 부족함이 식별될 경우, 이를 해결하기 위해 추가 변수를 개발하거나, 데이터를 요청한다.

모델링
- 데이터를 분석하기 위한 모델링 기법을 시행
- 프로세스 및 자원에 대한 제약이 있고 입력 값이 확률분포를 가지는 경우, 시뮬레이션 기법을 적용
- 프로세스 및 자원에 대한 제약이 있고 입력 값이 상수 값을 가지는 경우, 최적화 기법을 적용
- train 데이터와 test 데이터의 비율은 상황에 맞게 실시(6:4,7:3,8:2 등등)
- 과대적합, 과소적합이 되지 않도록 주의한다

모델 성능평가
- 데이터 마이닝 : 정확도, 정밀도, 시간
- 시뮬레이션 : Throughput, 평균 소요시간, 평균진행정도, 총 걸린 시간
- 최적화 : 최적화 전과 최적화 후의 값의 차이, 시간


### 3. 검증 및 테스트

#### 만든 모델이 얼마나 비즈니스적 영향력이 있는지 평가

- 실제 테스트에서는 분석 데이터(train)가 아닌 신규 데이터(test)에 모델을 적용하여 결과를 도출한다
- IT 업계에서 Unit Test에 해당된다

운영 상황에서의 테스트
- 분석 결과를 통해서 운영에 이를 반영할 수 있는지 연계성을 검증하는 단계
- 돌발 상황 및 outlier들 또한 문제 없이 모델링을 적용할 수 있는지 또한 검증함

고려사항
- 성능 테스트가 일관적으로 나오는지 확인
- 과대적합 및 과소적합이 발생하는 여부를 확인하고 발생할 경우 이를 수정한다
- 성능 테스트는 최소 3회 이상, 해당 기간은 일주일 이상을 권고한다.

비즈니스 영향도 평가
- 총 소유비용, 투자대비효과,순현재가치, 내부 수익률, 투자 회수 기간 등에 대한 평가


### 4. 적용

#### 한 모델이 영원히 동일한 성과를 낼 수 없기 때문에, 성과를 지속적으로 모니터링하여 이를 주기적으로 리모델링하는 업무

- 분석 결과를 업무 프로세스에 완전히 통합하여 실제 기간 단위(일,주,월)로 운영하는 것

운영 시스템에 적용 및 자동화
- 운영 시스템에 배치 스케쥴러 등을 통해 주기별로 분석 모델에 대한 성과가 나오고 있는지 모니터링을 하고, 이에 따른 성과자료를 누적
- 모니터링간 이상 현상 탐지 및 실제 데이터에 대한 모델링 결과를 배포
- 적용한 모델링 및 프로세스와 관련된 사항을들 문서화하여 이해 관계자들 간 공유

주기적 리모델링(유지보수)
- 새로운 모델에 대한 성과 모니터링을 지속적으로 시행하고, 일정 수준 내로 편차가 지속되도록 리모델링을 수행하여야 한다.
- 리모델링 시 수행 업무 : 데이터 마이닝, 시뮬레이션, 최적화 
1. 데이터 마이닝 : 최신 데이터를 이용한 재학습 및 변수 추가
2. 시뮬레이션 : 패턴 변화, 시간지연의 변화 등
3. 최적화 : 제약값의 변화 및 추가, 계수 변경과 같은 요소들에 대한 확인 및 재조정
