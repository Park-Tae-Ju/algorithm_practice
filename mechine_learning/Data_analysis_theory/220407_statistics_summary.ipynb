# 통계

### 1. 통계가 뭔지, 자료를 어떻게 획득하는지
- 통계: 특정 집단을 대상으로 수행한 조사나, 실험을 통해 나온 결과에 대한 요약된 형태의 표현, 또는 결과
- 필요한 자료: 조사 또는 실험을 통해 확보하고, 조사 대상에 따라 총조사(전수 조사)와 표본조사로 나뉨

통계자료의 획득 방법 관련 개념
- 모집단 : 조사하고자 하는 대상 집단 전체
- 원소 : 모집단을 구성하는 개체
- 표본 : 조사하기 위해 추출한 모집단의 일부 원소
- 모수 : 표본 관측에 의해 구하고자 하는 모집단에 대한 정보
- 통계적인 방법을 사용한다고 한다면, 모집단의 정의, 표본의 크기, 조사 방법, 조사 기간, 표본 추출 방법을 명확히 명시

표본 추출 방법
1. 단순 랜덤 추출(복원, 비복원 추출)
2. 계통 추출법 : 구간을 나눈 다음 각 구간 별로 항목을 추출
3. 집락 추출법 : 군집을 구분한 다음, 각 군집별로 단순 랜덤 추출을 한 후 모든 자료를 이용하거나 일부를 추출하여 사용(지역 표본 추출, 다단계표본추출)
4. 층화 추출법 : 이질적인 원소들로 이루어진 모집단에서 각 계측을 고루 대표할 수 있도록 표본을 추출하는 방법, 유사한 원소들을 몇개의 층으로 나누어 각 층에서 랜덤 추출하는 방법(비례층화추출법, 불비례층화추출법)

측정
- 데이터들을 주어진 목적에 적합하도록 측정하여 자료를 얻는 방법
1. 명목척도 : 측정 대상이 어떤 집단에 속하는지 분류할 때(성별, 지역)
2. 순서척도 : 측정 대상의 서열 관계를 관측하는 척도(만족도, 선호도, 학년, 신용등급)
3. 구간척도 : 측정 대상이 갖고 있는 속성의 양을 측정하는 것으로, 구간 별 간격이 의미가 있는 자료(온도, 지수)
4. 비율척도 : 간격에 대한 비유가 의미를 가지는 자료이며, 절대적인 기준인 0이 존재하고, 사칙연산이 가능한 척도 

### 2. 통계분석과 통계 분석 방법

통계분석의 정의
1. 기술 통계 : 주어진 자료로부터 판단, 예측과 같은 주관적인 과정을 배제하고 통계 집단의 특성을 수량화하여 객관적인 데이터로 나타내는 방법(평균, 표준편차, 중위수, 최빈값, 그래프)
2. 통계적 추론 : 수집된 자료를 이용해서 대상 집단(모집단)에 대한 의사결정을 하는 것 
- 모수 추정 : 표본집단을 통해 얻은 정보를 이용하여 모집단의 특징인 모수를 추론하는 것
- 가설 검정 : 대상 집단에 대해 특정한 가설을 설정한 후, 그 가설의 채택여부를 결정하는 것
- 예측 : 미래의 불확실성을 해결하기 위해 수행하는 것

### 3. 확률 및 확률분포

확률 : 특정 사건이 일어날 가능성의 척도
- 표본 공간 : 어떤 실험을 실시했을 때 나타날 수 있는 모든 결과들의 집합
- 사건 : 관찰자가 관심이 있는 사건으로 표본 공간의 부분집합
- 원소 : 나타날 수 있는 각각의 결과들
- P(E) = n(E) / n(omega) : 사건의 E의 확률은 표본공간의 원소의 갯수에 대한 사건 E의 개수의 비율

확률변수(random variable)
- 특정값이 나타날 가능성이 확률적으로 주어지는 변수
- 정의역(x)이 표본공간, 치역(y)이 실수값인 함수
- 0이 아닌 확률을 가지는 실수값의 형태에 따라 이산형 확률변수, 연속형 확률변수로 구분됨
1. 이산형 확률 변수 : 0이 아닌 확률값을 갖는 실수 값이 셀 수 있는 경우
- 베르누이 확률분포:
- 이항분포
- 기하분포
- 다항분포
- 포아송분포
2. 연속형 확률변수 : 가능한 값이 실수의 어느 특정 구간 전체에 해당하는 확률 변수(확률밀도함수)
- 균일분포
- 정규분포
- t-분포
- x^2 분포
- F-분포
