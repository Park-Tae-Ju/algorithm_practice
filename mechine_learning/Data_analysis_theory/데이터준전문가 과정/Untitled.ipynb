### 데이터의 정의
- 데이터는 객관적 사실이라는 의미이다.
- 개별 데이터로는 의미가 중요하지 않은 객관적인 사실을 나타낸다.
- 추론, 추측 등을 위한 근거로서 기능한다 =  상호 관계에서 가치를 갖는다

### 데이터의 유형
- 정성적 데이터 : 수치로 표현되지 않는 언어, 문장 등 ex) 수입이 늘었다
- 정량적 데이터 : 수치, 도형, 기호 등 ex) 우리집의 수입이 200만원 늘었다 

### 지식 경영 핵심 이슈
- 암묵지 : 경험을 통해 학습할 수 있는 지식 
- 형식지 : 문서 및 매뉴얼 등 체계화된 자료로 학습할 수 있는 지식

### 데이터와 정보의 관계
- DIKW 피라미드
- 계층구조 : 
1. Data(기호, 수치)
2. information(상관 관계를 통한 의미)
3. knowledge(예측 결과)
4. wisdom(아이디어)

## 데이터 베이스 정의와 특징

### 데이터 베이스 정의
- 데이터를 받아들이고 저장, 공급하기 위해 일정한 구조에 따라 편성된 데이터의 집합
- DBMS, DB라고도 불림

### 데이터 베이스 특징 및 특성
- 데이터베이스가 디지털로 구현된다는 특징을 이해하고 있는 것이 중요!

### 데이터베이스 발전과정
- 1세대(네트워크) - 2세대(관계형 DBMS) - 3세대(객체지향) - 4세대(NoSQL)

### 데이터베이스 설계 절차
- 요구사항 분석 - 개념 설계 - 논리 설계 - 물리적 설계

### 데이터웨어하우스
- 데이터베이스에서 필요한 정보를 추출하여 업무분석을 위한 하위 집합 형태의 데이터베이스
- SQL : 데이터베이스만을 위한 프로그래밍 언어(데이터 정의, 데이터 조작, 데이터 제어)

### 데이터 베이스 활용
- 현대 사회에서는 기업 전체를 관리하기 위해 데이터베이스를 사용하고 있음
- 고객 관리, 물류 관리, 지식 관리 등등을 하기 위함

# 데이터의 가치와 미래

### 빅데이터의 정의
- 규모, 다양성, 속도
- 진실성, 정확성, 휘발성, 가치

### 출현 배경
- 데이터 생성이 많아짐

### 빅데이터의 기능
- 산업혁명의 석탄, 철, 원유, 렌즈, 플랫폼 등에 비유됨

### 빅데이터의 본질적인 변화
- 데이터를 사후처리하여 새로운 정보를 얻어냄
- 표본조사가 아닌 전수조사
- 질보다 양
- 인과관계보다는 상관관계 분석을 통한 정보 습득

### 빅데이터의 가치
- 기존에 못했던 분석방법을 가능하게 함으로서 의미가 없었던 데이터들의 가치를 창출

### 빅데이터의 영향
- 정부, 기업 : 기존의 없었던 가치를 새롭게 얻어냄
- 개인 : 해당 사항이 드문 편

### 비즈니스 모델
- 활용 테크닉
1. 연관 규칙 학습 : 상관관계를 찾아냄
2. 유형분석 : 사용자를 분류함
3. 유전 알고리즘 : 문제의 해결책을 최적화
4. 기계학습 : 예측
5. 회귀분석 : 독립변수의 변화를 이용하여 종속변수가 얼마나 변화하는지를 확인(변수간 관계)
6. 감정분석 : 의견 분석 
7. 소셜 네트워크 분석 : 사용자간의 관계 및 영향력 파악

### 위기방안과 통제방안
1. 사생활 침해 / 개인정보 사용자의 책임 강화
2. 책임 원칙 훼손 / 기존의 책임 원칙 강화
3. 데이터 오용(잘못된 예측) / 알고리즘의 접근권 보장

- 읽어보기
