```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
#주피터 노트북에서만 쓰이는 기능
#노트북 내부에 그림을 표시하는 기능
%matplotlib inline
```

```python
#꺾은선 그래프 = line plot
#기본적으로 그래프를 그릴 때는 y축의 데이터만 있어도 되지만, x축의 범위를 지정해주고 싶을 때는 y축의 갯수에 맞게 넣어줘야한다.
plt.plot([10,20,30,40,50],[1,9,4,16,25]) #[(10,1),(20,9),(30,4),(40,16),(50,25)]
plt.show()
```

```python
#스타일 변경(색깔, 마커, 선종류)
#색깔 : https://matplotlib.org/2.0.2/examples/color/named_colors.html (빨간색 : r, 초록색 : g, 파란색 : b, 검정색 : k)
#마커 : (.) , (,) , (o) , (v,^,<,>), (1,2,3), (s) 등
#선종류 : (-),(--),(-.),(:)
plt.plot([10,20,30,40,50],[1,9,4,16,25],'b') 
plt.show()
```

```python
#스타일 변경(색깔, 마커, 선종류)
#세부 스타일 변경
#색깔 : color, c 
#선 굵기  :  linewidth , lw
#선 스타일 : linestyle, ls / example: (-) , (--) , (:) 등등
#점 : marker/ exmaple : (.), (,), (v,<,>,^)  등등
#점 크기 : markersize, ms
#점의 테두리 색깔  : markeredgecolor, mec
#점의 테두리의 굵기 : markeredgewidth, mew
#점의 내부 색깔 : markerfacecolor = mfc  

plt.plot([10,20,30,40,50],[1,9,4,16,25], c = 'b', lw = 1, 
         ls = '--', marker = '.', ms = 15, mec = 'r', mew = 3, mfc = 'k') 
plt.show()
```

```python
#x축과 y축의 범위 주기
#사용 용도 1. 데이터를 일부만 표기하고 싶을 때(특정 부분을 확대할 때)
#사용 용도 2. 데이터의 전체 형태를 표기하고 싶을 때()
plt.plot([10,20,30,40,50],[1,9,4,16,25], c = 'b', lw = 1, 
         ls = '--', marker = '.', ms = 15, mec = 'r', mew = 3, mfc = 'k')

plt.xlim(0,30)
plt.ylim(0,15)

plt.show()
```

```python
#사용 용도 2. 데이터의 전체 형태를 표기하고 싶을 때()
plt.plot([20,30,40],[9,4,16], c = 'b', lw = 1, 
         ls = '--', marker = '.', ms = 15, mec = 'r', mew = 3, mfc = 'k')

#데이터의 범위를 더 크게 주기
plt.xlim(0,50)
plt.ylim(0,30)

plt.show()
```

```python
#축의 값을 표시 : tick
#수식을 넣고 싶다면 Latex를 검색해서 수식을 표기하는 방법을 참조
plt.plot([20,30,40],[9,4,16], c = 'b', lw = 1, 
         ls = '--', marker = '.', ms = 15, mec = 'r', mew = 3, mfc = 'k')

plt.xticks([15,25,35,45])
plt.yticks([5,10,15])
```

```python
#화면에 축의 위치를 명확하게 하기 위한 점선을 표기 : grid
#모눈종이처럼 화면에 명시
plt.plot([20,30,40],[9,4,16], c = 'b', lw = 1, 
         ls = '--', marker = '.', ms = 15, mec = 'r', mew = 3, mfc = 'k')
plt.grid(True, zorder = 0)
plt.show()
```

```python
#여러개의 선 그리기(1)
plt.plot([20,30,40],[9,4,16],'g.:'
        ,[30,40,50],[1,9,35],'rs--')
```

```python
#여러개의 선 그리기(2)
plt.plot([20,30,40],[9,4,16], c = 'b', lw = 1, 
         ls = '--', marker = '.', ms = 15, mec = 'r', mew = 3, mfc = 'k')
plt.plot([1,20,60],[5,25,45], c = 'g', lw = 1, 
         ls = ':', marker = 's', ms = 10, mec = 'k', mew = 3, mfc = 'k')
plt.show()
```

```python
#범례 만들기
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.plot(X,C, '-', label = 'cosine') #코사인 플롯
plt.plot(X,S, ':', label = 'sine') #사인 플롯

plt.legend(loc=0) #0 : best, 나머지는 1~9범위

# plt.grid()
plt.show()
```

```python
#라벨, 제목 넣기

X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.plot(X,C, '-', label = 'cosine') #코사인 플롯
plt.plot(X,S, ':', label = 'sine') #사인 플롯

plt.xlabel('X_value')
plt.ylabel('cos_and_sin')
plt.title('plot_name')

plt.grid()
plt.show()
```

```python
#subplot 그리기
#matlab 라이브러리의 인덱스를 기준으로 정렬됨
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

ax1 = plt.subplot(2,1,1) # 가로 : 2, 세로 1, 내가 정한 위치가 1
plt.plot(X,C, 'r-', label = 'cosine') #코사인 플롯
plt.grid()

ax2 = plt.subplot(2,1,2)
plt.plot(X,S, 'g:', label = 'sine') #사인 플롯
plt.grid()

plt.show()
```

```python
#subplot 그리기
#matlab 라이브러리의 인덱스를 기준으로 정렬됨
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

ax1 = plt.subplot(2,2,1) # 가로 : 2, 세로 1, 내가 정한 위치가 1
plt.plot(X,C, 'r-', label = 'cosine') #코사인 플롯
plt.grid()

ax2 = plt.subplot(2,2,2)
plt.plot(X,S, 'g:', label = 'sine') #사인 플롯
plt.grid()

ax3 = plt.subplot(2,2,3) # 가로 : 2, 세로 1, 내가 정한 위치가 1
plt.plot(X,C, 'k-', label = 'cosine') #코사인 플롯
plt.grid()

ax4 = plt.subplot(2,2,4)
plt.plot(X,S, 'b:', label = 'sine') #사인 플롯
plt.grid()

plt.show()
```

### 여러가지 차트 또는 플롯

1. 막대 그래프(bar)
2. stem plot
3. 원형 그래프(pie)
4. 히스토그램
5. 점 그래프(scatter)
6. 이미지 그래프(ex : 진할수록 값이 높음)
7. 등고선 그래프
8. 3차원 그래프

```python
#bar chart
x = np.arange(3)
y = [2,5,1]
xlabel = ['a','b','c']

plt.bar(x,y,zorder = 10) #zorder = 레이어의 우선순위
plt.xticks(x,xlabel)
plt.grid(alpha = 0.3, zorder = 1)
plt.show()
```

```python
#stem plot
x = np.linspace(0.1, 3*np.pi, 15)
y = np.cos(x)
plt.stem(x,y,'r-.')
plt.show()
```

```python
#원형 그래프
#원형 그래프 그릴 때 : 전체 중 일부의 퍼센테이지 표기 등
labels = ['A','B','C','D','E']
sizes = [50,15,13,15,7]
plt.pie(sizes, labels=labels)
plt.show()
```

```python
#원형 그래프 세부 설정하기
labels = ['A','B','C','D','E']
sizes = [50,15,13,15,7]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'm']
explode = (0.1,0,0.1,0,0) #간격주기
 
plt.pie(sizes, labels=labels, colors = colors, explode = explode,autopct = '%1.1f%%',shadow = True)
plt.show()
```

```python
#히스토그램
x = np.random.randn(1000)
arrays, bins, patchs = plt.hist(x, bins = 10)
plt.show()

print(bins) # x축에 해당하는 데이터의 범주
print(arrays) # x축(범주)에 해당하는 데이터의 갯수
```

```python
#등고선 그래프 그리기
#3차원 데이터일때 씀(x,y,z : 가로,세로,높이)

def data(x,y) :
    return (1 - x / 2 + x**5 + y ** 3) * np.exp(-x**2 - y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
XX, YY = np.meshgrid(x,y) #좌표
ZZ = data(XX,YY) #높이

plt.contourf(XX,YY,ZZ, alpha = 0.5, cmap = 'jet')
plt.contour(XX,YY,ZZ, colors = 'k')
plt.grid()
plt.show()
```

```python
#3차원 서피스 플롯(등고선과 비슷)

from mpl_toolkits.mplot3d import Axes3D
x = np.arange(-4,4.25)
y = np.arange(-4,4.25)
XX,YY = np.meshgrid(x,y) #직사각형 격자를 만들어주는 함수
ZZ = np.sin(np.sqrt(XX**2 + YY**2))

#3D로 화면을 그리기
fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(XX,YY,ZZ, rstride = 1, cstride = 1, cmap = 'summer', antialiased=True)
plt.show()

```
