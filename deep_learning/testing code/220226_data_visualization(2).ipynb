```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#데이터 불러오기(seaborn)

iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
#커널 함수를 이용한 그래프
#확률 밀도 함수 중 하나이며 커널함수를 이용해서, 히스토그램의 문제점을 개선한 방법
#커널 함수를 겹치는 방법으로 분포를 부드러운 곡선의 모습으로 보여줌.

x = iris.petal_width.values
sns.kdeplot(x)
plt.show()
```

```python
#distplot
sns.displot(x, kde=True)
plt.show()
```

```python
#countplot
#카테고리별 값이 얼마나 있는지 알 수 있는 그래프
#DataFrame만 가능함

sns.countplot(x='class', data = titanic)
plt.show()
```

```python
sns.countplot(x = 'day', data = tips)
plt.show()
```

```python
#실수형 데이터만 있을 때 쓸 수 있는 그래프
#scatter 그래프

#꽃받침의 길이와 넓이
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris)
plt.show()
```

```python
#실수형 데이터만 있을 때 쓸 수 있는 그래프
#kind에 따라서 그래프의 형태가 바뀜

#꽃받침의 길이와 넓이
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'kde')
plt.show()
```

```python
#여러개의 칼럼 간 데이터 시각화
# ex) 1,2,3,4 칼럼이 있다고 할 때, 자기자신을 포함하여 그래프를 그리므로 4x4 그래프가 그려짐
# 데이터에 칼럼이 너무 많을 경우에는 어떻게 해야될까? = pandas
sns.pairplot(iris)
plt.show()
```

```python
#데이터가 카데고리 - 카테고리인 경우 heatmap이 유효하다
titanic_size = titanic.pivot_table(index = 'class', columns= 'sex', aggfunc = 'size') #aggsize = size = 갯수
titanic_size
```

```python
sns.heatmap(titanic_size)
```

```python
sns.heatmap(titanic_size, cmap=sns.light_palette('gray'),annot=True, fmt='d') #fmt(format), d = 십진수
plt.show()
```

```python
#모든 데이터 분석, 예측은 통계적인 지식을 기반으로 함
#모집단(전체집단)의 분포가 표본집단(우리가 조사한 데이터)의 분포와 유사하거나 동일하다는 것을 가정으로 함
#표본집단의 특징을 확인할 수 있으면, 이것은 모집단도 동일하거나 유사한 특징을 가지고 있을 것이므로
#새롭게 들어오는 데이터 또한 동일한 특징, 즉 동일한 분포 또는 카테고리, 실수 값 등을 가질 것이라고 할 수 있다.
#여기까지가 통계

#데이터 시각화는 통계적인 내용을 기반으로 사람들에게 설명하는 것이므로
#통계적인 개념들이 주된 경우가 많음
```

```python
#실수-카데고리가 섞여있는 데이터를 시각화하기
#1. bar 차트
sns.barplot(x='day', y = 'total_bill', data = tips)
plt.grid()
plt.show()
```

```python
#2. boxplot
#각 박스는 1분위수 ~ 3분위수 범위를 뜻한다. 즉 전체 데이터의 50%
#박스 가운데의 선은 중위값, 중앙값이다(평균과 다름)
#박스 위 아래의 선은 전체 데이터의 범위를 나타낸다
#박스와 선 바깥에 있는 점은 이상치, 이상점이며, outlier라고도 한다.(일반적인 사례에서 벗어난 경우)
sns.boxplot(x='day', y = 'total_bill', data = tips)
plt.show()
```

```python
#3. violinplot
# boxplot과 유사하지만, 실제적인 분포와 조금 더 유사한 형태, 커널 밀도 함수와 유사하다.
sns.violinplot(x='day', y = 'total_bill', data = tips, zorder = 5)
plt.zorder_(zorder = 1)
plt.show()
```

```python
#3. stripplot
# scatter과 유사하지만, 카테고리와 실수의 조합을 표현할 수 있음
# 단점 : 점이 겹치는 경우가 있음
sns.stripplot(x='day', y = 'total_bill', data = tips, zorder = 5, size = 3)
plt.grid(zorder = 1)
plt.show()
```

```python
#4. swarmplot
# stripplot과는 다르게 가로로 확장하여 데이터가 덜 겹침
sns.swarmplot(x='day', y = 'total_bill', data = tips, zorder = 5)
plt.grid(zorder = 1)
plt.show()
```

```python
#다차원 데이터를 시각화하기
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips)
plt.show()
```

```python
#boxplot
sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips)
plt.show()
```

```python
#violinplot
sns.violinplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips)
plt.grid()
plt.show()
```

```python
#stripplot
sns.stripplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips)
plt.grid()
plt.legend(loc=0)
plt.show()
```

```python
#swarmplot
sns.swarmplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips)
plt.grid()
plt.legend(loc=0)
plt.show()
```

```python
#시각화 방법 변경
sns.violinplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips, split = True)
plt.legend(loc = 2)
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips, dodge = True)
plt.grid()
plt.legend(loc=0)
plt.show()
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips, dodge = True, size = 4)
plt.grid()
plt.legend(loc=0)
plt.show()
```

```python
#다차원 데이터를 heatmap으로 나타낼 경우
flights_data = flights.pivot('month','year','passengers')
sns.heatmap(flights_data, fmt = 'd', annot = True, linewidths=1)
plt.show()
```

```python
#각 그래프를 합쳐보기
#1. box와 strip plot 겹쳐보기
sns.boxplot(x = 'tip', y = 'day', data = tips, orient= 'h')
sns.stripplot(x = 'tip', y = 'day', data = tips, color = '0.7')
plt.show()
```

```python
sns.swarmplot(x = 'day', y = 'total_bill',  data = tips, color = '0.6')
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```
