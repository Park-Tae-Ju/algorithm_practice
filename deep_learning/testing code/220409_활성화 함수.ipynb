# Activation function

데이터에서 수많은 특징,특성,feature들을 통해 데이터를 분류할 수 있는데  
이 데이터 중 필요있는 데이터와 필요없는 데이터를 분류하여 데이터 폭을 정의하는 역할

#### 최초의 신경망 모델 = 퍼셉트론
- a = w1x1 + w2x2+w3x3 .... + b
- 여기서 y = f(a)에서 f(x)에 해당되는 부분이 활성화 함수

#### 계단 함수(step)

- 최초의 활성화 함수
- 데이터의 손실이 쉽게 일어날 가능성이 크다

#### 시그모이드(sigmoid) 함수

- 계단함수와 다르게 값의 결과가 실수 값으로 나오는 활성화 함수
- f(x)의 값이 아무리 커도 0~1사이의 값으로 나오기 때문에 미분할 경우 그 값이 점점 0에 수렴한다
- 값의 중심이 0.5이기 때문에 문제점이 발생한다.

#### 하이퍼볼릭탄젠트 함수

- 시그모이드의 문제점을 일부 해결한 활성화 함수
- 값의 중심이 0이므로 얻은 값의 결과를 유지할 수 있음 
- 값의 범위가 -1 ~ 1이므로 음수인 경우도 해결이 가능
- 기울기가 소실되는 문제는 해결할 수 없었음

####  RELU 함수

- sigmoid 함수와 하이퍼볼릭탄젠트의 단점(그래디언트의 소실)을 해결한 함수
- 성능(연산)도 뛰어난 편

#### softmax 함수
- softmax는 결과값이 확률로 나오는 함수
- 결과를 분류로 해야되는 경우 softmax를 쓴다.

#### elu

- relu의 0을 출력함으로 인해 학습이 안되는 문제를 방지하기 위해 만들어진 개념
- 0 이하의 값이 있다고 하더라도 이를 일정 값으로 유지함
- 0 이하일 때 비선형함수 형태로 나타남

#### leak relu

- relu의 0을 출력함으로 인해 학습이 안되는 문제를 방지하기 위해 만들어진 개념
- 0 이하의 값이 있다고 하더라도 이를 일정 값으로 유지함
- 0 이하일 때 선형함수 형태로 나타남

### 활성화 함수를 쓰는 이유
- 기울기 소실(그래디언트 베니싱 문제)를 보완하기 위함이다.
- 처음부터 기울기가 0인 경우에는 처리가 불가능하다

#### Tip
- relu를 쓰는 게 무난함
- 만약 최종적인 결과가 분류를 해야되는 경우에는 softmax를 씀
- relu를 쓰는 경우가 성능을 도출해야되는데 성능이 다소 애매한 경우 leak relu, elu를 시도
