# loss?
컴퓨터가 대입한 결과와 실제 정답간의 차이를 loss라고 함
딥러닝 및 머신러닝의 학습은 이 loss를 줄이는 방향으로 학습이 진행이 됨

### 1.  MSE (Mean Squared Error)
- 예측한 값과 실제값의 오차를 제곱해서 평균한 것
- 제곱으로 인해 오차가 양수든 음수든 손실함수의 누적 값을 증가시킴


### 2. RMSE (Root Mean Squared Error)
- 예측한 값과 실제값의 오차를 제곱해서 평균한 값을 제곱근한 것
- 오차를 제곱함으로 인해 너무 커진 값을 제곱근을 통해 크기를 다시 줄임 = 값의 왜곡을 줄임


### 3. CrossEntropy
- 예측 분포와 실제 분포를 비교했을 때 그 차이에 대한 것의 평균
- 2개의 클래스가 있을 때 쓰면 좋다



### 4. Categorical CrossEntropy
- 클래스가 2개 이상일 때 쓰면 좋다
- softmax 다음에 연계되어 나오는 경우가 잦음
