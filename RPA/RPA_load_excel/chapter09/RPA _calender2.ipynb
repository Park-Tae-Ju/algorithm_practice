```python
from datetime import datetime
import pandas as pd

import itertools

from pytimekr import pytimekr

from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.styles import Alignment, PatternFill, Font, Side, Border, numbers
```

```python
#날짜 리스트 생성
def date_range(start, end):
    start = datetime.strptime(start, "%Y%m%d")
    end = datetime.strptime(end, "%Y%m%d")
    dates = [date.strftime("%m.%d") for date in pandas.date_range(start, periods=(end-start).days+1)]
    dates = [dates[i:i+7] for i in range(0,len(dates),7)]
    return dates
    

startDay = input('수업 시작 날짜를 입력해주세요(20220101) : ')
endDay = input('수업 종료 날짜를 입력해주세요(20220101) : ')

```

```python
dates = date_range(startDay, endDay)
len(dates)
```

```python
#달력 만들기
dates = date_range(startDay, endDay)
#병합할 리스트 만들기
checkClass = [[''] * 7] * len(dates)

#합치기
temp = list(itertools.chain(*zip(dates, checkClass)))
calenderClass = pd.DataFrame(temp)
temp = pd.DataFrame({None: ['날짜', '수업']*(calenderClass.index.stop // 2)}) #인덱스 이름을 None으로 줘서 안나오게
calenderClass = pd.concat([calenderClass, temp], axis = 1)

#범례 수정
calenderClass.set_index(None, inplace = True) 
calenderClass.columns = ['일','월','화','수','목','금','토']
```

```python
wb = Workbook()
ws = wb['Sheet']
ws.title = "생성한 스케쥴"

r_idx,c_idx = 0,0 # 열처리
rows = dataframe_to_rows(calenderClass)
for r_idx, row in enumerate(rows, 4):
    for c_idx, value in enumerate(row, 2):
        ws.cell(row=r_idx, column=c_idx, value=value)
        r_idx,c_idx = r_idx,c_idx
        
if len(dates) <= 5:
    ws['B4'] = '평일반 스케쥴'
else:
    ws['B4'] = '주말반 스케쥴'
ws['B{}'.format(r_idx+1)] = '국가공휴일은 휴강'
ws.merge_cells('B4:I5') #평일반 스케쥴
ws.merge_cells('B{}'.format(r_idx+1)+':'+'I{}'.format(r_idx+1)) #국가 공휴일            

#위 아래 사이즈 및 볼드처리
ws['B4'].alignment = Alignment(horizontal='center', vertical= 'center')
ws['B4'].font = Font(sz = "15", bold = True)
ws['B{}'.format(r_idx+1)].alignment = Alignment(horizontal='right', vertical= 'center')
ws['B{}'.format(r_idx+1)].font = Font(sz = "11", bold = True)


# wb.save('new_Class_calender_{}_{}.xlsx'.format(startDay, endDay))

# calenderClass.to_excel('new_Class_calender_{}_{}.xlsx'.format(startDay, endDay), startcol=1, startrow=5)


```
